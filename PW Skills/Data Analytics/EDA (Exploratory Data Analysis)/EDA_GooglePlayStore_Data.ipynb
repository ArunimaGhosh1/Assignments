###**Convert "M" to "k" and replace "M" and "k".**

```python
df_copy['Size']
```

```python
df_copy['Size'].unique()
```

####1 Mb = 1024 Kb

```python
item = '19M'
item[-1]
```

```python
item.replace('M', '')
```

```python
def size_process(item):
  if str(item)[-1] == 'M':
    res = float(str(item).replace('M', ''))
    res = res*1024
    return res

  elif str(item)[-1] == 'k':
    res = float(str(item).replace('k', ''))
    return res

  else:
    return str(np.NaN)
```

```python
df_copy['Size'] = df_copy['Size'].apply(size_process)
df_copy['Size']
```

##Q4. What are top 5 installed category apps in each type?
Category & Type dono me groupy karke Installs ka sum (groupby on Category & Type columns with aggregation on Installs column)
Ans. Game of free type.

```python
df_copy.groupby(['Type','Category'])['Installs'].sum().reset_index().sort_values(by=['Type', 'Installs'], ascending=[True, False]).groupby('Type').head(5).reset_index(drop=True)
```

```python
top_5_installs = df_copy.groupby(['Category', 'Type'])['Installs'].sum().sort_values(ascending=False).reset_index()[:5]
top_5_installs
```

```python
plt.figure(figsize=(7,4))
sns.barplot(top_5_installs, x=top_5_installs['Category'], y=top_5_installs['Installs'], palette='Set2')
plt.xticks(rotation=15)
```

```python
def top_5_installs(df_copy):
  grouped_df = df_copy.groupby(['Type','Category'])['Installs'].sum().reset_index()
  grouped_df = grouped_df.sort_values(by=['Type','Installs'], ascending=[True, False])
  top_5 = grouped_df.groupby('Type').head(5).reset_index(drop=True)

  return top_5

top_5_installs(df_copy)
```

###Ans.- Category Game of Type Free is highest installed.
