####1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

####2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

````python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

output_file('scatterplot.html')
# Step 1: Generate random data
np.random.seed(42)  # for reproducibility
n_points = 100
x = np.random.random(size=n_points) * 100
y = np.random.random(size=n_points) * 100

# Step 2: Generate random sizes for the markers
sizes = np.random.random(size=n_points) * 30 + 10  # sizes between 10 and 40

# Step 3: Generate random colors for the markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n_points)

# Prepare data source
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Step 4: Create a Bokeh scatter plot
output_notebook()

p = figure(title="Random Scatter Plot with Varying Sizes and Colors",
           x_axis_label='X Axis', y_axis_label='Y Axis')

p.scatter(x='x', y='y', size='sizes', color='colors', alpha=0.6, source=source)

show(p)
````

####3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
####```fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']```
####```counts = [20, 25, 30, 35]```

````python
output_file('bar.html')

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, height=350)
p.vbar(x=fruits, top=counts, width=0.9)

p.y_range.start=0
show(p)
````

````python
output_file('bar1.html')

from bokeh.models import ColumnDataSource
from bokeh.palettes import Bright6
from bokeh.plotting import figure, show
from bokeh.transform import factor_cmap

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, height=350, title="Fruit Counts", toolbar_location=None)
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field='fruits', fill_color=factor_cmap('fruits', palette=Bright6, factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start=0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
````

####4. Create a Bokeh histogram to visualize the distribution of the given data.

#### 5. Create a Bokeh heatmap using the provided dataset.
####```data_heatmap = np.random.rand(10, 10)```
####```x = np.linspace(0, 1, 10)```
####```y = np.linspace(0, 1, 10)```
####```xx, yy = np.meshgrid(x, y)```

````python
output_file('heatmap.html')

from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import linear_cmap
from bokeh.palettes import Bright6
from bokeh.transform import factor_cmap

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

data = {'x':xx.flatten(), 'y':yy.flatten(), 'value':data_heatmap.flatten()}
df = pd.DataFrame(data)

mapper = LinearColorMapper(palette='Bright6', low=df.value.min(), high=df.value.max())

p = figure(title="Data Heatmap", tools="hover", toolbar_location=None, height=500, width=500)

p.rect(x='x', y='y', source=df, width=1/len(x), height=1/len(y), line_color=None, fill_color=linear_cmap('value', palette='Bright6', low=df.value.min(), high=df.value.max()))

color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
````
