Here are three different methods for creating identical 2D arrays in NumPy, along with explanations and code examples:

**Method 1:** Using `numpy.full` with a manually filled array

**Method 2:** Using `numpy.ones` and then modifying the array:

````python
# Creating a 2D array filled with ones and then setting the desired values
array2 = np.ones((2, 3), dtype=int)
array2[0] = [1, 2, 3]
array2[1] = [4, 5, 6]

print("Method 2 - Array using numpy.ones and then setting values:\n", array2)
````

**Method 3:** Using `numpy.zeros` and then modifying the array:

````python
# Creating a 2D array filled with zeros and then setting the desired values
array3 = np.zeros((2, 3), dtype=int)
array3[0] = [1, 2, 3]
array3[1] = [4, 5, 6]

print("Method 3 - Array using numpy.zeros and then setting values: \n", array3)
````

####**3. Explain the following terms:**

- The difference in `np.array`, `np.asarray` and `np.asanyarray`.

In NumPy, `np.array`, `np.asarray`, and `np.asanyarray` are functions used to convert input data into NumPy arrays, but they have slight differences in behavior:

1. `np.array`:
   - This function creates a new NumPy array object from the input data.
   - If the input data is already a NumPy array, it creates a copy by default.
   - You can control whether a copy is made using the copy parameter (copy=True or copy=False).
   - Example:

````python
a = [1, 2, 3]
b = np.array(a)  # Creates a new array from the list
c = np.array(b)  # Creates a copy of the array b
id(b), id(c)
````

2. `np.asarray`:
   - This function converts the input data to a NumPy array, but it does not create a copy if the input is already a NumPy array (or array-like object).
   - It is more efficient when you are sure that the input is already an array and you do not need a copy.
   - Example:

````python
a = [1, 2, 3]
b = np.asarray(a)  # Creates a new array from the list
c = np.asarray(b)  # Does not create a copy, c is the same array as b
id(b), id(c)
````

3. `np.asanyarray`:
   - Similar to np.asarray, but it is more flexible with subclasses of ndarray.
   - If the input data is an instance of a subclass of ndarray (e.g., matrix), np.asanyarray will return the input as is, without converting it to a base ndarray.
   - Useful when working with arrays that might be of a subclass type and you want to preserve that type.
   - Example:

````python
a = [1, 2, 3]
b = np.asanyarray(a)  # Creates a new array from the list
c = np.asanyarray(b)  # Does not create a copy, c is the same array as b
d = np.matrix([[1, 2], [3, 4]])
e = np.asanyarray(d)  # Returns the same matrix object, not converting to ndarray
````

In summary, use `np.array` when you need a new array or want to ensure a copy is made, `np.asarray` when you want to avoid unnecessary copying and are working with array-like objects, and `np.asanyarray` when you want to avoid copying and also preserve subclasses of ndarray.

- The difference between Deep copy and shallow copy.

A deep copy and a shallow copy are both ways to duplicate objects in Python, but they differ in how they handle the objects being copied, especially when it comes to nested objects (objects that contain other objects like lists or dictionaries).

### Shallow Copy
A shallow copy creates a new object, but it does not create copies of the objects contained within the original object. Instead, it copies references to the objects within the original object. As a result, both the original and the copied object share references to the same nested objects.

- *Implications*: Modifications to nested objects in the copied object will affect the original object and vice versa.

- *Example*:

````python
import copy
original = [[1, 2, 3], [4, 5, 6]]
shallow_copy = copy.copy(original)

shallow_copy[0][0] = 'X'

print(original)       # Output: [['X', 2, 3], [4, 5, 6]]
print(shallow_copy)   # Output: [['X', 2, 3], [4, 5, 6]]
````

### Deep Copy
A deep copy creates a new object as well as new copies of all objects found in the original object, recursively. This means that the copied object and the original object are completely independent of each other.

- *Implications*: Modifications to nested objects in the copied object will not affect the original object and vice versa.

- *Example*:

````python
original = [[1, 2, 3], [4, 5, 6]]
deep_copy = copy.deepcopy(original)

deep_copy[0][0] = 'X'

print(original)      # Output: [[1, 2, 3], [4, 5, 6]]
print(deep_copy)     # Output: [['X', 2, 3], [4, 5, 6]]
````

### Summary
- *Shallow Copy*:
  - Creates a new object.
  - Copies references to the nested objects.
  - Changes to nested objects affect both the original and the copied object.
  - Created using copy.copy() or slicing (new_list = old_list[:]).

- *Deep Copy*:
  - Creates a new object.
  - Recursively copies all nested objects.
  - Changes to nested objects do not affect the original object.
  - Created using copy.deepcopy().

####**15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:**

####a) Plot a histogram of the data with 30 bins.
####b) Overlay a line plot representing the normal distribution's probability density function (PDF).
####c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
####d) Set the title of the plot as 'Histogram with PDF Overlay'.

````python
from scipy import stats

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a histogram with 30 bins
plt.hist(data, bins=30, density=True, color='red', label='Histogram')

# Calculate the PDF of the normal distribution
x = np.linspace(min(data), max(data), 100)
pdf = stats.norm.pdf(x, loc=0, scale=1)

# Overlay the PDF as a line plot
plt.plot(x, pdf, color='blue', label='PDF')

# Label the axes and set the title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Show the plot
plt.legend()
plt.show()
````

####**17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.**

````python
import seaborn as sns

# Generate two random arrays
x1 = np.random.randn(1000)
y1 = np.random.randn(1000)

# Determine the quadrant for each point
quadrant = np.where(x1 >= 0, np.where(y1 >= 0, 1, 4), np.where(y1 >= 0, 2, 3))

# Create a scatter plot with Seaborn, coloring points based on quadrant
sns.scatterplot(x=x1, y=y1, hue=quadrant, palette='viridis')

# Add a legend
plt.legend(title='Quadrant')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
````

####**18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.**

````python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# This line ensures that the plot will be displayed in the notebook
output_notebook()

# Generate data for the sine wave
x2 = np.linspace(0, 4 * np.pi, 100)  # x values
y2 = np.sin(x2)  # y values, sine of x

# Create a figure object
p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='sin(X)', width=800, height=400)

# Add a line renderer with legend and line thickness
p.line(x2, y2, legend_label="Sine Wave", line_width=2)

# Add grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

# Show the plot
show(p)
````

####**19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.**

````python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Spectral6

# Display the plot in the notebook
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E', 'F']
values = [np.random.randint(1, 100) for cat in categories]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a figure object
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values',
           frame_height=400, frame_width=600, tools="")

# Add a bar renderer with a color map based on values
p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color=factor_cmap('categories', palette=Spectral6, factors=categories))

# Add hover tooltips to display exact values
hover = HoverTool()
hover.tooltips = [
    ("Category", "@categories"),
    ("Value", "@values"),
]
p.add_tools(hover)

# Show the plot
show(p)
````
